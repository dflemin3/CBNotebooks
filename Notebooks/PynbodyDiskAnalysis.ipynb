```python
%matplotlib inline
#Imports required to run

import pynbody
from matplotlib.pylab import *
import matplotlib.pylab as plt
import pynbody.plot.sph as sph

#Typical plot parameters that make for pretty plots
plt.rcParams['figure.figsize'] = (10,6)
plt.rcParams['font.size'] = 18
```

```python
#Load Snapshot created using Isaac's code
#Remember to have snapshot AND param file in same directory as this notebook
s = pynbody.load('snapshot.000750')
```

```python
print s.stars[0]['pos']
```

```python
#Convert to physical units
s.physical_units()
#Now convert to units I want (Au instead of kpc, keep Msol)
s.gas['pos'].convert_units('au')
s.gas['rho'].convert_units('Msol au**-3')
#Check to make sure it worked
print s.gas['rho'].units
print s.gas['pos'].units
#Note: type s.gas['something'].in_units('new units') to convert units temporarily
```

```python
#Create profile, check units
#Check units
p = pynbody.analysis.profile.Profile(s,max='20 au',min='0.01 au',type='log',nbins=100)
pg = pynbody.analysis.profile.Profile(s.gas,max='20 au',nbins=100)
```

```python
#Try plotting stuff
#Plot the gas surface density as a function of radius for test disk

#Now do the plotting with data derived from gas profile, pg
#In this case, tangential velocity vs. radius
plt.plot(pg['rbins'], pg['density'],'k', label = '$Gas$',)
plt.legend()
#plt.semilogy()
plt.xlabel("$R$ [AU]")
plt.ylabel(r'$\Sigma$ [M$_{\odot}$ au$^{-2}$]')
plt.show()
#Print units to confirm
print pg['rbins'].units
print pg['density'].units
```

```python
#Try plotting temperature profile of gas in plane of disk
#T vs. radius
plt.plot(pg['rbins'], pg['temp'],'k', label = '$Gas$')
plt.legend()
#plt.semilogy()
plt.xlabel("$R$ [AU]")
plt.ylabel("$T$ [K]")
plt.show()
```

```python
#Plot sound speed vs. radius in plane of disk
print pg['cs'].units
print pg['rbins'].units
plt.plot(pg['rbins'],pg['cs'],'k', label='$Gas$')
plt.xlabel("$R$ [AU]")
plt.ylabel("$C_s [km s^{-1}]$")
plt.legend()
plt.show()
```
