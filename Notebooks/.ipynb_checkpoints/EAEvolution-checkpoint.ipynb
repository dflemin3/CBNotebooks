#Notes:
I think the difference in phase between the sink/no-sink oscillation is stochastic in nature.  I imagine that the spiral arms of the gas disk form stochastically with random initial locations such that for either case, the binary stars are in a different phase of the orbit.  Given this, the two different formation locations will lead to different tugs on the stars such that although the eccentricity oscillation will have pretty much the same amplitude, its phase can be off.

```python
plt.plot(timeNS,semiNS,'b',label='No Sinks')
plt.plot(time,semi,'g',label='Sinks')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel("Time (years)")
plt.ylabel("Semimajor Axis (AU)")
plt.title("Semimajor Axis vs. Time Comparison CW")
```
